```python
import numpy as np
np.__version__
```

**Norma Wektora, Długość Wektora W R^2**

```python
v1 = np.array([-1, 4])
v1
```

```python
np.linalg.norm(v1)
```

**Norma Euklidesowa W R^n**

```python
v2 = np.array([-10, 10, 13])
np.linalg.norm(v2)
```

**Odległóść Dwóch Punktów Na Płaszczyźnie**

```python
p = np.array([3, 0])
q = np.array([0, 4])

np.linalg.norm(p - q)
```

**Odległóść Dwóch Punktów W Przestrzeni R^n**

```python
p = np.array([3, 0, 1])
q = np.array([0, 4, 2])

np.linalg.norm(p - q)
```
